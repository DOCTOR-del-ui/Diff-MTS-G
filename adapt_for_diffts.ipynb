```python
import os
import numpy as np
import pandas as pd
```

```python
modelname = "DiffUnet"
submodelname = "mtsplus"
```

```python
for idx in range(1, 5):
    for window in [96]:
        save_dir = os.path.join("AdaptDiffTS", "OUTPUT", f"cmapsst{submodelname}{idx}_{window}")
        os.makedirs(save_dir, exist_ok=True)

        output_dir = os.path.join("weights", "syn_data")
        
        diffmts_datapath = os.path.join(output_dir, f"syn_FD00{idx}_{modelname}_{window}ddpm.npz")
        diffmts_gendata = np.load(diffmts_datapath)  # 转为 float32
        diffmts_gendata_npy = diffmts_gendata["data"].astype(np.float64)
        
        print(f"================FD00{idx}-{window}-start================")

        print(f"=================FD00{idx}-{window}-end=================")
        print("...")
        
        ts_savepath = os.path.join(save_dir, f"ddpm_fake_cmapsst{submodelname}{idx}_{window}.npy")
        np.save(ts_savepath, diffmts_gendata_npy)
```
